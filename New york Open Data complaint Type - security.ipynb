```python
from urllib.request import urlopen
import json
import pandas as pd
url="https://nycopendata.socrata.com/resource/erm2-nwe9.json?$limit=50000"

response = urlopen(url)
####data_json=json.loads(response.read())
json_data=response.read().decode('utf-8','replace')
d=json.loads(json_data)
##df=pd.(d[1])
data_list=[]
df=pd.DataFrame(columns=['unique_key','created_date','agency','agency_name','complaint_type'])
    
for datados in d:
    unique_key=datados['unique_key']
    created_date=datados['created_date']
    agency=datados['agency']
    agency_name=datados['agency_name']
    complaint_type=datados['complaint_type']
    
    
    data_item={
            'unique_key':unique_key,
            'created_date':created_date,
            'agency':agency,
            'agency_name':agency_name,
            'complaint_type':complaint_type,
    }
    data_list.append(data_item)
 
    
for i in range(0,len(data_list)):
    df.loc[i]=[data_list[i]['unique_key'],data_list[i]['created_date'],data_list[i]['agency'],data_list[i]['agency_name'],data_list[i]['complaint_type']]
##df    
    ##df.columns=['unique_key','created_date','agency','agency_name']
    ##print(df.head())
##for i in d:  

##for chunk in df:
  ##      categories=['unique_key','created_date','agency','agency_name']
    ##    details=chunk[categories]   
      ##  display(details.head())
        ##break
import numpy as np
import  matplotlib.pyplot as plt

##plt.style.use("fivethirtyeight")
##df_complaint=df["complaint_type"].tail(7)
##df_complaint.value_counts().plot(kind="pie",
##                            autopct='%1.1f%%', # add in percentages
##                            startangle=90,     
##                            shadow=True)       # add shadow   )
##plt.show()

dt_pie=df["complaint_type"].value_counts().head(15)

##dt_pie.columns=['complaint_type','Count_complaint']

dt_pie.plot(kind="pie",autopct='%1.1f%%')
# add in percentages
##                            startangle=90,     
##                            shadow=True)       # add shadow   )
##plt.show()
plt.title('The Top 7 complaint type to Present by New york Open Data')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```
